```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T = 200
U = 20  # Number of different u values

# Generate U evenly spaced values of u in [0,1]
u_values = np.linspace(0, 1, U)
phi = lambda u: 0.2 + 0.3 * np.sin(0.1 * u)

# Generate the functional time series for each u value
Y = np.zeros((U, T))
Y[:, 0] = np.random.normal(0, 1, U)  # Initialize the series for each u value

for u_index, u in enumerate(u_values):
    for t in range(1, T):
        Y[u_index, t] = phi(u) * Y[u_index, t-1] + np.random.normal(0, 1)
        
# 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert the 2D time series into 3D coordinates for plotting
time, U_mesh = np.meshgrid(np.arange(T), u_values)

ax.plot_surface(time, U_mesh, Y, cmap='viridis', linewidth=0.5)
#cmap be may changed to plasma, inferno,magma, cividis, coolwarm, spring, or Blues
ax.set_xlabel('Time')
ax.set_xlabel('Time')
ax.set_ylabel('u')
ax.set_zlabel('Y(t,T)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T = 200
U = 20  # Number of different u values

# Generate U evenly spaced values of u in [0,1]
u_values = np.linspace(0, 1, U)

phi_1 = lambda u: 0.1 + 0.3 * np.sin(0.2 * u)
phi_2 = lambda u: -0.4 + 0.2 * np.cos(0.05 * u)

# Generate the functional time series for each u value
Y = np.zeros((U, T))
Y[:, 0:2] = np.random.normal(0, 1, (U, 2))  # Initialize the series for each u value

for u_index, u in enumerate(u_values):
    for t in range(2, T):
        Y[u_index, t] = phi_1(u) * Y[u_index, t-1] + phi_2(u) * Y[u_index, t-2] + np.random.normal(0, 1)

# 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Convert the 2D time series into 3D coordinates for plotting
time, U_mesh = np.meshgrid(np.arange(T), u_values)

# Change the style
ax.plot_surface(time, U_mesh, Y, cmap='viridis', linewidth=0.5)
#cmap be may changed to plasma, inferno,magma, cividis, coolwarm, spring, or Blues
ax.set_xlabel('Time')
ax.set_ylabel('u')
ax.set_zlabel('Y(t,T)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T = 200
phi = lambda u: 0.2 + 0.3 * np.sin(0.1 * u)

# Generate the time series
Y = np.zeros(T)
Y[0] = np.random.normal(0, 1)  # Initialize the series

for t in range(1, T):
    u = t / T
    Y[t] = phi(u) * Y[t-1] + np.random.normal(0, 1)

# Plotting
plt.figure(figsize=(12, 8))
plt.plot(np.arange(T), Y, label='Y_{t,T}', color='blue')
plt.xlabel('Time')
plt.ylabel('Y_{t,T}')
plt.title('tvAR(1) Time Series with u = t/T')
plt.legend()
plt.grid(True)
plt.show()
```
