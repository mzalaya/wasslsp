```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functional form for phi. 
def phi(t, x, i):
    if i == 1:
        return 0.6 + 0.2 * np.sin(t/T) * x
    if i == 2:
        return 0.3 - 0.1 * np.cos(t/T) * x

# Parameters
T = 1000  # number of time points
p = 2   # order of the AR model

# Initialize the series
y = np.zeros(T)
x = np.sin(np.linspace(0, 10, T))  # Functional covariate: sine wave for illustration

# Generate the TVAR series
for t in range(p, T):
    for i in range(1, p + 1):
        y[t] += phi(t, x[t], i) * y[t - i]
    y[t] += np.random.normal()

# 3D Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

t_values = np.arange(T)

ax.plot(t_values, x, y)
ax.set_xlabel('Time')
ax.set_ylabel('Functional Covariate X_t')
ax.set_zlabel('Time Series Y_t')
ax.set_title('3D Plot of TVAR Series with Functional Covariate')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functional form for phi. 
def phi(t, x, i):
    if i == 1:
        return 0.6 + 0.2 * np.sin(t) * x
    if i == 2:
        return 0.3 - 0.1 * np.cos(t) * x

# Parameters
T = 1000  # number of time points
p = 2   # order of the AR model

# Initialize the series
y = np.zeros(T)
x = np.sin(np.linspace(0, 10, T))  # Functional covariate: sine wave for illustration

# Generate the TVAR series
for t in range(p, T):
    for i in range(1, p + 1):
        y[t] += phi(t, x[t], i) * y[t - i]
    y[t] += np.random.normal()

# 3D Plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

t_values = np.arange(T)

ax.plot(t_values, x, y, lw=1)
ax.set_xlabel('Time')
ax.set_ylabel('Functional Covariate X_t')
ax.set_zlabel('Time Series Y_t')
ax.set_title('3D Plot of TVAR Series with Functional Covariate')

plt.show()

# Set the background color of the plot to white
ax.patch.set_facecolor('white')  # Set outer color to white
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))  # White x-axis background
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))  # White y-axis background
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))  # White z-axis background

# Also set the grid color to a light gray for better contrast against the white background
ax.xaxis.pane.set_edgecolor('gray')
ax.yaxis.pane.set_edgecolor('gray')
ax.zaxis.pane.set_edgecolor('gray')
ax.grid(True, color='gray', linestyle=':', linewidth=0.5)

# Change the perspective
elevation_angle = 45  # can be adjusted
azimuthal_angle = 45  # can be adjusted
ax.view_init(elevation_angle, azimuthal_angle)

plt.show()
```
