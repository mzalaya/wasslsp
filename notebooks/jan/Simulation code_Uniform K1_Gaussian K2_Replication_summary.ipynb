```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_excel('F:\Jan_Lenovo\My Documents\PhD in UTC\Research\Drafts\Figures\Example 2\W1_Replications_50_uniform and gaussian.xlsx')

data = [df['T200'], df['T300'], df['T400'],df['T500']]
meanprop = dict(marker='^', markeredgecolor='green', markerfacecolor='green')

plt.figure(figsize=(6, 4))
bp=plt.boxplot(data, showmeans=True, meanprops=meanprop)
#plt.title("1-Wassertein distances of each T for 50 replications")
plt.xlabel("Data length (T)")
plt.ylabel("1-Wasserstein distance")
plt.xticks([1, 2, 3, 4], ['200', '300', '400', '500'])
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
# Labeling the mean values
for i, line in enumerate(bp['means']):
    mean_val = np.mean(data[i])
    plt.annotate(f'{mean_val:.6f}', 
                 (line.get_xdata()[0], line.get_ydata()[0]), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center', 
                 fontsize=6, 
                 color='red')

plt.show()
```
