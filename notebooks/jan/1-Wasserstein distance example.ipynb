```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define two Gaussian distributions
mu1, sigma1 = 0, 0.5
mu2, sigma2 = 1, 0.5

# Generate sample points
x = np.linspace(-3, 3, 1000)

# Compute the PDFs
pdf1 = norm.pdf(x, mu1, sigma1)
pdf2 = norm.pdf(x, mu2, sigma2)

# Compute the CDFs
cdf1 = norm.cdf(x, mu1, sigma1)
cdf2 = norm.cdf(x, mu2, sigma2)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, pdf1, label='PDF P', color='blue')
plt.plot(x, pdf2, label='PDF Q', color='red')
plt.fill_between(x, pdf1, pdf2, where=(pdf1 > pdf2), interpolate=True, color='blue', alpha=0.3)
plt.fill_between(x, pdf1, pdf2, where=(pdf1 <= pdf2), interpolate=True, color='red', alpha=0.3)
plt.title('(a) Probability Density Functions')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, cdf1, label='CDF P', color='blue')
plt.plot(x, cdf2, label='CDF Q', color='red')
plt.fill_between(x, cdf1, cdf2, where=(cdf1 > cdf2), interpolate=True, color='blue', alpha=0.3)
plt.fill_between(x, cdf1, cdf2, where=(cdf1 <= cdf2), interpolate=True, color='red', alpha=0.3)
plt.title('(b) Cumulative Distribution Functions')
plt.legend()

plt.tight_layout()
plt.show()
```
